# Benchmarking Midcaps over Large caps

Sometimes midcaps give more returns than large-caps and sometimes it is the other way around. What happens if you benchmark a mid-cap index with a large-cap index as a reference? A study.

```r
library(DBI)
library(plutoDbR)
library(plutoR)
library(tidyverse)
options("scipen"=999)
options(stringsAsFactors = FALSE)
source("/usr/share/pluto/config.R")
source("/usr/share/pluto/goofy/plot.common.R")
source("/usr/share/pluto/goofy/misc.common.R")

library(ggpubr)
library(ggthemes)
library(quantmod)
library(PerformanceAnalytics)

options(repr.plot.width=16, repr.plot.height=8)

indices <- Indices()
```

**Note**: these indices and dates are obtained from the [60/40](https://github.com/stockviz/plutons/blob/master/60-40-equity-bond-india.R.ipynb) notebook

```r
portfolioIndex <- "NIFTY MIDCAP 100 TR"
benchmarkIndex <- "NIFTY 50 TR"
riskfreeIndex <- "0_5"
startDate <- as.Date("2004-01-01")
endDate <- as.Date("2019-08-31")
shortCodes <- c("PORT", "BENCH", "RF")
lookBack <- 220 #days ~ 1 year
```

```r
port <- indices$NseTimeSeries() %>%
    filter(NAME == portfolioIndex & TIME_STAMP >= startDate & TIME_STAMP <= endDate) %>%
    select(TIME_STAMP, CLOSE) %>%
    collect() %>%
    mutate(TIME_STAMP = as.Date(TIME_STAMP))

bench <- indices$NseTimeSeries() %>%
    filter(NAME == benchmarkIndex & TIME_STAMP >= startDate & TIME_STAMP <= endDate) %>%
    select(TIME_STAMP, CLOSE) %>%
    collect() %>%
    mutate(TIME_STAMP = as.Date(TIME_STAMP))

riskfree <- indices$IndiaGsecTimeSeries() %>%
    filter(NAME == riskfreeIndex & TIME_STAMP >= startDate & TIME_STAMP <= endDate) %>%
    select(TIME_STAMP, CLOSE = TRI) %>%
    collect() %>%
    mutate(TIME_STAMP = as.Date(TIME_STAMP))

allXts <- merge(xts(port$CLOSE, port$TIME_STAMP), xts(bench$CLOSE, bench$TIME_STAMP), xts(riskfree$CLOSE, riskfree$TIME_STAMP))
names(allXts) <- shortCodes
allXts$RF <- na.locf(allXts$RF)

allXts <- na.omit(allXts)

allRets <- merge(dailyReturn(allXts[,1]), dailyReturn(allXts[,2]), dailyReturn(allXts[,3]))
names(allRets) <- shortCodes
```

```r
#always inspect the data
print(head(allRets))
print(tail(allRets))
```

### Cumulative returns

```r
Common.PlotCumReturns(allRets, sprintf("%s vs. %s; rf = %s", portfolioIndex, benchmarkIndex, riskfreeIndex), "total return")
```

There isn't much in "excess" returns that jumps out from a cumulative point-of-view. On an annualized basis, mid-caps barely give a point of excess over large-caps.

We now calculate rolling sharpe, alpha and information ratio.
By setting the calculation period as 220 days, we get the approximate annual values.

CAPM.alpha(Ra, Rb, Rf) from PerformanceAnalytics [docs](https://www.rdocumentation.org/packages/PerformanceAnalytics/versions/1.5.2/topics/CAPM.alpha)
SharpeRatio(R, Rf, FUN) from PerformanceAnalytics [docs](https://www.rdocumentation.org/packages/PerformanceAnalytics/versions/1.5.2/topics/SharpeRatio)
InformationRatio(Ra, Rb) from PerformanceAnalytics [docs](https://www.rdocumentation.org/packages/PerformanceAnalytics/versions/1.5.2/topics/InformationRatio)

```r
alphas <- rollapply(allRets, lookBack, function(X) CAPM.alpha(X$PORT, X$BENCH, X$RF), by.column = F)
traditionalSharpe <- rollapply(allRets, lookBack, function(X) SharpeRatio(X$PORT, X$RF, FUN = "StdDev"), by.column = F)   
ir <- rollapply(allRets, lookBack, function(X) InformationRatio(X$PORT, X$BENCH), by.column = F)      
                
allMetrics<-na.omit(merge(alphas, traditionalSharpe, ir))               
```

```r
#always inspect the data
print("ALPHA")
print(head(alphas))
print(tail(alphas))

print("Traditional Sharpe")
print(head(traditionalSharpe))
print(tail(traditionalSharpe))

print("IR")
print(head(ir))
print(tail(ir))

print("MERGED")
print(head(allMetrics))
print(tail(allMetrics))
```

Plot the metrics.

```r
metricsDf<-data.frame(allMetrics)
names(metricsDf)<-c('ALPHA', 'SHARPE', 'IR')

metricsDf$T<-index(allMetrics)

print(head(metricsDf))
print(tail(metricsDf))

firstDate<-first(index(allMetrics))
lastDate<-last(index(allMetrics))
xAxisTicks<-seq(from=firstDate, to=lastDate, length.out=10)
```

```r
plot2 <- metricsDf %>%
    select(T, ALPHA) %>%
    ggplot() +
    geom_line(aes(x = T, y = ALPHA), size = 0.5, alpha = 0.75) +
    ylab("alpha") +
    theme_economist() +
    theme(axis.title.x = element_blank(),
        axis.text.x=element_blank(),
        axis.ticks.x=element_blank())

plot3 <- metricsDf %>%
    select(T, SHARPE) %>%
    ggplot() +
    geom_line(aes(x = T, y = SHARPE), size = 0.5, alpha = 0.75) +
    ylab("sharpe") +
    theme_economist() +
    scale_x_date(breaks = xAxisTicks) +
    theme(axis.title.x = element_blank(),
        axis.text.x=element_blank(),
        axis.ticks.x=element_blank())

plot4 <- metricsDf %>%
    select(T, IR) %>%
    ggplot() +
    geom_line(aes(x = T, y = IR), size = 0.5, alpha = 0.75) +
    ylab("ir") +
    theme_economist() +
    scale_x_date(breaks = xAxisTicks) +
    theme(axis.title.x = element_blank()) +
    annotate("text", x=lastDate, y=min(metricsDf$ir), label = "@StockViz", hjust=1.1, vjust=-1.1, col="white", cex=6, fontface = "bold", alpha = 0.8)
    
mergedPlot <- rbind(ggplotGrob(plot2), ggplotGrob(plot3), ggplotGrob(plot4), size = "first")
annotate_figure(mergedPlot, top = text_grob(sprintf("%s vs. %s; rf = %s", portfolioIndex, benchmarkIndex, riskfreeIndex), face = "bold", size = 14, family='Segoe UI'))
```

There is nothing in these charts that pound the table for mid-caps over large-caps.
